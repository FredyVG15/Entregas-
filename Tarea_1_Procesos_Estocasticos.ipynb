**CAMINATA ALEATORIA EN 1D**

**Descrpción del problema:**  Consideremos que tenemos una partícula con una probabilidad $q$, dezplazandose una unidad a la izquierda y una probabilidad $p$,  moviendose con una unidad a la derecha.
De tal manera que:  $p$ + $q$ = $1$.




```python
import matplotlib.pyplot as plt #para generacion de gráficas
import random as r #para generación de numeros aleatorios
```

```python
def caminata(p, N): #Definimos una función llamada caminata, con los parámetros p= probabilidad de moverse a la derecha y N= Número total de pasos
  l = 0 # Inicia la posición en 0, la caminata comienza en el origen
  posiciones = [0] # lista que guarda todas las posiciones
  for i in range(N): # Ciclo en donde cada repetición represenra un paso
    if r.random() < p: #Para generar un número aleatorio entre 0 y 1, si ese némero es < que p, entonces ocurre evento con probabilidad p
      l = l + 1 # La particula se mueve a la derecha, aumenta la su posición un 1
    else:
      l = l - 1 # Si la particula no cumple condición se mueve a la izquierda, disminuye en -1
    posiciones.append(l) #Para contruir la trayectoria completa debido a que guarda la nueva posición en la lista
  return posiciones

```

```python
#Para uso de la función
p = 0.5 #probabilidad de moverse a la derecha
N = 100 # Número total de pasos
posiciones = caminata(p, N) #para ejecución de la funcion, guarda la trayectoria generada en la variable posiciones
```

```python
plt.plot(posiciones)# Dibuja la trayectoria
plt.title("Caminata Aleatoria")        # Título agregado
plt.axhline(y=0, color='black', linestyle='--')  # línea punteada
plt.xlabel("pasos")
plt.ylabel("posición")
plt.show()# Muestra la gáfica
```

**ANÁLISIS Y CONCLUSIONES**

Se modeló una caminata aleatoria donde:

- La partícula inicia en el origen:

$$
X_0 = 0
$$

- En cada paso:

Se mueve $+1$ con probabilidad $p$:

$$
P(\Delta X = +1) = p
$$

Se mueve $-1$ con probabilidad:

$$
q = 1 - p
$$

$$
P(\Delta X = -1) = 1 - p
$$

- Se simulan $N$ pasos.

- La posición en el paso $n$ está dada por:

$$
X_n = \sum_{i=1}^{n} \Delta X_i
$$

donde

$$
\Delta X_i =
\begin{cases}
+1 & \text{con probabilidad } p \\
-1 & \text{con probabilidad } 1-p
\end{cases}
$$

El objetivo fue analizar el comportamiento del proceso dependiendo del valor de $p$.

Ante lo anteriormente presentado, ¿qué preguntas pueden surgir de este fenómeno?

En primera instancia, nos concentraremos en el valor de probabilidad $p$, ¿qué pasa si lo hacemos variar?,¿qué sucede con la media y la varianza?

•	Si $p=0.5$, se observa que la trayectoria no muestra uan  preferencia: sube y baja sin mostrar un sesgo particular.

•	Si $p>0.5$, se observa que tiende a irse hacia la derecha, es decir, su trayectoria toma valores positivos.

•	Si $p<0.5$, se observa que su trayectoria tiende a irse hacia la izquierda, es decir, toma valores negativos.

Por tanto, aunque exista una tendencia, siempre habrá variación porque los pasos son aleatorios.




¿Cuál es la posición esperada después de $n$ pasos?
Entonces, iniciamos con:

$$
\mathbb{E}[Y_i] = (+1)p + (-1)(1 - p)
$$

$$
\mathbb{E}[Y_i] = p - (1 - p)
$$

$$
\mathbb{E}[Y_i] = 2p - 1
$$

Como los pasos son independientes:

$$
\mathbb{E}[X_n] = n\,\mathbb{E}[Y_i]
$$

$$
\boxed{
\mathbb{E}[X_n] = n(2p - 1)
}
$$

Por lo tanto, se pude interpretar de la siguiente forma:

- Si $p = 0.5 \Rightarrow \mathbb{E}[X_n] = 0$
- Si $p > 0.5 \Rightarrow$ tendencia positiva
- Si $p < 0.5 \Rightarrow$ tendencia negativa


¿Cuél es la varianza de la posición?

 Varianza de la Caminata Aleatoria

Primero calculamos la varianza de un paso:

$$
\operatorname{Var}(Y_i) = \mathbb{E}[Y_i^2] - (\mathbb{E}[Y_i])^2
$$

Como $Y_i^2 = 1$:

$$
\operatorname{Var}(Y_i) = 1 - (2p - 1)^2
$$

$$
\operatorname{Var}(Y_i) = 4p(1 - p)
$$

Como los pasos son independientes:

$$
\boxed{
\operatorname{Var}(X_n) = 4np(1 - p)
}
$$

Es decir, que la varianza crece linealmente con $n$.

Entoces, si $p$=$0.5$ se presenta que $$
\mathbb{E}[X_n] = 0\
$$ y
$$
\operatorname{Var}(X_n) = n
$$


Por otro lado, si $p \neq 0.5$, se presenta que
$$
\mathbb{E}[X_n] = n(2p - 1)
$$, es decir, existe una tendencia y en cuanto a la varianza:
$$
\operatorname{Var}(X_n) = 4np(1 - p)
$$
es decir, sigue creciendo linealmente.



Otra interrogante es ¿Se pude predecir el comportamiento promedio?, ¿qué pasa cuando $n$ es grande?

La respouesta es sí debido a que

$$
\frac{X_n}{n} \longrightarrow 2p - 1
$$
lo anterior por la Ley de los grandes números.



Teorema Central del Límite

$$
\frac{X_n - n(2p - 1)}{\sqrt{4np(1 - p)}}
\longrightarrow
\mathcal{N}(0,1)
$$

Cuando $n$ es grande, la distribución se aproxima a una normal.




Convergencia al Movimiento Browniano

Si escalamos adecuadamente:

$$
\frac{X_n}{\sqrt{n}}
$$

converge a un proceso continuo llamado Movimiento Browniano.

**CONCLUSIÓN**

La caminata aleatoria muestra que, aunque cada trayectoria individual es impredecible, su comportamiento promedio sigue una estructura matemática clara.

La media crece linealmente con el número de pasos cuando existe tendencia,es decir, $p \neq 0.5$ mientras que la dispersión aumenta proporcionalmente a $$\sqrt{n}.$$

Este modelo conecta el mundo discreto con el continuo (Movimiento Browniano) y es base de aplicaciones importantes en física y finanzas.

En concreto, la caminata aleatoria demuestra cómo del azar puede surgir un comportamiento estadísticamente ordenado.

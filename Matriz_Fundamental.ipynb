***1.- ¿Cuál es el número promedio de tiradas necesarias para terminar el juego de serpientes y escaleras en el tablero de la imagen adjunta?***

\[
\begin{array}{|c|c|c|c|}
\hline
17\downarrow 9 & 18 & 19 & 20\\
\hline
16 & 15\uparrow 19 & 14 & 13\downarrow 4\\
\hline
9 & 10 & 11 & 12\\
\hline
8 & 7 & 6 & 5\\
\hline
1 & 2 & 3\uparrow 11 & 4\\
\hline
\end{array}
\]


**Solución:**


Para resolver de manera analítica el número promedio de tiradas necesarias para terminar este juego, se modelará el tablero como una **Cadena de Markov Absorbente**.

Asumiremos las reglas estándar:
* Se utiliza un dado justo de 6 caras ($D6$).
* El jugador comienza fuera del tablero (casilla 0).
* Cualquier tirada que alcance o supere la casilla 20 resulta en la victoria (estado absorbente).

## Definición del Espacio de Estados y Transiciones
El sistema tiene un espacio de estados $S = \{0, 1, 2, \dots, 20\}$.
El estado $20$ es nuestro **estado absorbente**, y los estados del $0$ al $19$ son **estados transitorios**.

Primero, mapeamos las alteraciones del tablero (serpientes y escaleras).

Cuando un jugador cae exactamente en el inicio de una de estas, se traslada instantáneamente a su destino.

Definimos la función de salto $D(x)$:

* **Escaleras:** $D(3) = 11$,  $D(15) = 19$
* **Serpientes:** $D(13) = 4$, $D(17) = 9$
* **Casillas normales:** $D(x) = x$
* **Meta:** Para $x \ge 20$, $D(x) = 20$

La idea analítica es convertir el juego en una cadena de Markov absorbente y calcular el tiempo medio hasta llegar al estado final.

Suponemos que el juego termina cuando el jugador llega o rebasa la casilla 20.

El estado inicial es 0, es decir, antes de entrar a la casilla 1.

## Formulación de las Ecuaciones de Esperanza (Ecuación del tiempo esperado)

Sea $E_i$ el número esperado de tiradas para llegar a la casilla 20 partiendo desde la casilla $i$. Nuestro objetivo es encontrar $E_0$.

Para la meta, el número de tiradas esperadas es cero:
$$E_{20} = 0$$

Para cualquier otra casilla $i$, el valor esperado es 1 tirada (la que estamos a punto de hacer) más el promedio de los valores esperados de las casillas en las que podemos aterrizar, ponderado por la probabilidad del dado ($1/6$):
$$E_i = 1 + \frac{1}{6} \sum_{d=1}^{6} E_{D(i+d)}$$

Como no podemos permanecer en las casillas de inicio de serpientes o escaleras, sus esperanzas son iguales a las de sus destinos:
$$E_3 = E_{11}, \quad E_{13} = E_4, \quad E_{15} = E_{19}, \quad E_{17} = E_9$$

PodemosSe puede construir el sistema de ecuaciones de atrás hacia adelante para ver su comportamiento analítico:

**Casilla 19 y 18:** Cualquier número del 1 al 6 hace que se alcance o supere el 20.
$$E_{19} = 1 + \frac{1}{6}(0 \cdot 6) = 1$$
$$E_{18} = 1$$

**Casilla 16:** Las tiradas posibles son 1 (cae en 17, baja por serpiente al 9), 2 (casilla 18), 3 (casilla 19), y 4, 5, 6 (ganan).
$$E_{16} = 1 + \frac{1}{6}(E_9 + E_{18} + E_{19} + E_{20} + E_{20} + E_{20})$$
$$E_{16} = 1 + \frac{1}{6}(E_9 + 1 + 1 + 0) = \frac{4}{3} + \frac{1}{6}E_9$$

**Casilla 14:** Las tiradas posibles son 1 (cae en 15, sube por escalera al 19), 2 (casilla 16), 3 (cae en 17, baja por serpiente al 9), 4 (casilla 18), 5 (casilla 19), 6 (gana).
$$E_{14} = 1 + \frac{1}{6}(E_{19} + E_{16} + E_9 + E_{18} + E_{19} + E_{20})$$
Sustituyendo los valores ya conocidos:
$$E_{14} = 1 + \frac{1}{6}\left(1 + \left(\frac{4}{3} + \frac{1}{6}E_9\right) + E_9 + 1 + 1 + 0\right) = \frac{31}{18} + \frac{7}{36}E_9$$

Se continua este proceso de sustitución hacia atrás generará un sistema lineal cerrado para las 16 variables independientes restantes.

## Solución Matricial (La Matriz Fundamental)

En la teoría de Cadenas de Markov, en lugar de sustituir algebraicamente 20 variables, construimos la matriz de transición por bloques:

$$P = \begin{pmatrix} Q & R \\ 0 & I \end{pmatrix}$$

Donde $Q$ es la matriz de $20 \times 20$ que contiene las probabilidades de transición entre los estados transitorios (0 al 19).

El vector que contiene el tiempo esperado hasta la absorción partiendo desde cada estado transitorio se calcula encontrando la **Matriz Fundamental** $N$:

$$N = (I - Q)^{-1}$$

El tiempo esperado $\mathbf{t}$ es:

$$\mathbf{t} = N \cdot \mathbf{1}$$

Donde $\mathbf{1}$ es un vector columna de unos. El valor de $E_0$ será la primera entrada de este vector $\mathbf{t}$.

Entonces,

Al resolver:

$$ t = (I - Q)^{-1} \mathbf{1} $$

se obtiene para el estado inicial 0:

$$ t_0 = \frac{271925032608}{1953165253573} $$

$$ t_0 \approx 7.1827 $$

Por lo tanto:

$$ E[T] \approx 7.18 $$
$$ E_0 \approx 7.1827 $$

Es decir, en promedio se necesitan aproximadamente 7 tiradas para terminar el juego.





**SOLUCIÓN ANALÍTICA**

```python
# Importamos la librería sympy
# Se utiliza para trabajar con matemáticas simbólicas
# como fracciones exactas, matrices e inversas
import sympy as sp

#Se crea matriz Q
# Número de estados transitorios
# Estados: 0,1,2,...,19
# la matriz Q (20x20) de estados transitorios (0 a 19)
# Q contiene probabilidades entre estados transitorios
n = 20
Q = sp.zeros(n, n)


# Diccionario:
# clave = casilla donde cae
# valor = casilla a la que se mueve
# Diccionario de saltos (Serpientes y Escaleras)
saltos = {3: 11, 15: 19, 13: 4, 17: 9}

# Llenar matriz de transición
for i in range(n): ## Recorre cada estado del tablero


    # Recorre los posibles resultados del dado
    # dado = 1,2,3,4,5,6
    for dado in range(1, 7):
        destino = i + dado # Nueva posición después de lanzar el dado

        # Si destino >= 20,es decir, si juego llega o rebasa la meta
        # el juego termina

        if destino >= 20:
            continue # Probabilidad va al estado absorbente (fuera de Q)

            #Si aun está en el tablero
        else:
            # Aplicar reglas de serpientes/escaleras
            destino_final = saltos.get(destino, destino)
            Q[i, destino_final] += sp.Rational(1, 6)

# Matriz Identidad
I = sp.eye(n) # Se crea la matriz identidad de tamaño 20x20

# Matriz Fundamental N = (I - Q)^-1
N = (I - Q).inv()

# Vector de unos
unos = sp.ones(n, 1) # Se  Crea un vector columna de unos:Tiene tamaño 20x1

# Calcular tiempos esperados
t = N * unos

# El resultado es el estado inicial 0
esperanza_0 = t[0]

print(f"Promedio exacto de tiradas: {esperanza_0}")
print(f"Es decir: {esperanza_0.evalf():.4f} tiradas")
```

**SOLUCIÓN POR SIMULACIÓN (MONTE CARLO)**

```python
# Librerias implementadas

import random # random permite generar números aleatorios
import numpy as np # numpy para calcular promedios
```

```python
#Considerando los datos de la imagen(tablero)

meta = 20 # Define la casilla final del juego. En este tablero, el juego termina al llegar o rebasar la casilla 20.

# Serpientes y escaleras, es decir, crea un diccionario con las serpientes y
# escaleras del tablero.
saltos = {3: 11, 15: 19, 13: 4, 17: 9}
```

```python
#Construccion de una Función que simula una partida#

def jugar(): #Define una función llamada jugar. Esta función representa una partida completa de serpientes y escaleras.

    # El jugador inicia en la casilla 0
    posicion = 0 #La casilla 0 representa el punto de inicio antes de entrar al tablero.

    # Contador de tiradas
    tiradas = 0 #Al inicio vale 0 porque todavía no se ha lanzado el dado.


    # Mientras no llegue a la meta, este ciclo se repite mientras la posición del jugador sea menor que 20.#
    while posicion < meta:

        # Generamos un dado aleatorio de 1 a 6
        dado = random.randint(1, 6)#Simula el lanzamiento de un dado.

        # Sumamos una tirada
        tiradas += 1 #Aumenta el contador de tiradas en 1

        # Avanzamos
        posicion += dado #Actualiza la posición del jugador.


        # Si llega o rebasa la meta,
        # termina el juego
        if posicion >= meta:#Verifica si el jugador llegó o rebasó la casilla final.
            break #detiene la partida porque el jugador ya terminó el juego.


        # Si cae en serpiente o escalera,
        # se mueve automáticamente
        if posicion in saltos:  #Revisa si la nueva posición está en el diccionario saltos. Esto significa que el jugador cayó en una casilla donde hay serpiente o escalera.
            posicion = saltos[posicion] #Actualiza la posición usando el destino correspondiente en el diccionario.


    # Regresamos el número total de tiradas
    return tiradas #cuando termina la partida, la función regresa el número total de tiradas usadas.
```

```python
##Para las Simulaciones

# Número de partidas simuladas
sim = 100000 #Define cuántas partidas se van a simular

# Lista donde guardaremos los resultados
resultados = [] #Aquí se guardará el número de tiradas que tomó cada partida.


# Ciclo en el que Repetimos el juego muchas veces
for i in range(sim):

    # Guardamos el número de tiradas
    resultados.append(jugar())
```

```python
##Para el promedio

# Calculamos el promedio de tiradas
promedio_simulado = np.mean(resultados) #Calcula el promedio de todos los valores guardados en resultados, este valor estima el número promedio de tiradas necesarias para terminar el juego.


print("Promedio simulado de tiradas:")
print(promedio_simulado)
```

***2.-¿Cuál es la probabilidad de que el ratón, iniciando  en la casilla 0, alcance la comida?***

$$
\begin{array}{|c\ \ c|c\ \ c|c|}
\hline
0 & & 1 & & 7\ \text{food} \\
\hline
2 & & 3 & & 4 \\
\hline
8\ \text{shock} & & 5 & & 6 \\
\hline
\end{array}
$$


**Solución:**

Estructuraremos el sistema utilizando la Matriz Fundamental. Asi, separamoa los estados de transición de los estados de absorción.

## Definición del Espacio de Estados
Clasificamos las 9 casillas del laberinto en dos conjuntos:

- **Estados Transitorios ($T$)**: $\{0, 1, 2, 3, 4, 5, 6\}$
- **Estados Absorbentes ($A$)**: $\{7 \text{ (Comida)}, 8 \text{ (Descarga)}\}$


Buscamos:

$$P(\text{llegar a comida antes que shock} \mid X_0 = 0)$$

## Conexiones del laberinto


- $0 \rightarrow \{1, 2\}$
- $1 \rightarrow \{0, 3, 7\}$
- $2 \rightarrow \{0, 3, 8\}$
- $3 \rightarrow \{1, 2, 4, 5\}$
- $4 \rightarrow \{3, 6, 7\}$
- $5 \rightarrow \{3, 6, 8\}$
- $6 \rightarrow \{4, 5\}$

Como el ratón elige al azar entre las salidas disponibles, cada salida tiene probabilidad $\frac{1}{k}$, donde $k$ es el número de salidas disponibles.



Bajo esta clasificación, la matriz de transición $P$ adquiere su forma canónica por bloques:

$$P = \begin{pmatrix} Q & R \\ 0 & I \end{pmatrix}$$

Donde $Q$ es la matriz de transiciones entre estados transitorios ($7 \times 7$) y $R$ es la matriz de transiciones de estados transitorios a absorbentes ($7 \times 2$).

##Construcción de las Matrices $Q$ y $R$
Observando las conexiones y puertas del laberinto, asumimos que en cada intersección el ratón elige con probabilidad uniforme.

**Matriz $Q$ ($7 \times 7$):**
$$Q = \begin{pmatrix}

$$Q = \begin{pmatrix}

0   & 1/4 & 1/4 & 0   & 1/4 & 1/4 & 0 \\
0   & 0   & 0   & 1/3 & 0   & 0   & 1/3 \\
0   & 0   & 0   & 1/3 & 0   & 0   & 1/3 \\
0   & 0   & 0   & 0   & 1/2 & 1/2 & 0
\end{pmatrix}$$


**Matriz $R$ ($7 \times 2$):**  
Las columnas corresponden a los estados 7 y 8. Solo desde las casillas 4 y 5 se puede acceder a estos estados directamente.
$$R = \begin{pmatrix}
0 & 0 \\

0 & 0 \\

\end{pmatrix}$$



##Planteamiento de la Matriz Fundamental $N$
La matriz fundamental se define como $N = (I - Q)^{-1}$.  
Cada elemento $n_{i,j}$ de esta matriz representa el número esperado de veces que el ratón visitará el estado transitorio $j$ dado que comenzó en el estado $i$.

Para encontrar la probabilidad de absorción en cualquier estado, utilizamos la matriz $B = N \cdot R$.

Queremos calcular la probabilidad de llegar a la comida (estado 7) iniciando en 0. Esto corresponde al elemento $b_{0,7}$ de la matriz $B$:
$$b_{0,7} = \sum_{k=0}^{6} n_{0,k} \cdot r_{k,7}$$

Observando la matriz $R$, el único valor no nulo en la columna del estado 7 es $r_{4,7} = 1/3$. Por lo tanto, la sumatoria se simplifica a:
$$b_{0,7} = n_{0,4} \cdot \frac{1}{3}$$

Esto significa que no necesitamos invertir toda la matriz $7 \times 7$. Solo necesitamos calcular $n_{0,4}$, que es un elemento de la primera fila de $N$.

## Cálculo Analítico de la Fila 0
Sea $\mathbf{n}_0 = (n_0, n_1, n_2, n_3, n_4, n_5, n_6)$ la primera fila de $N$. Por definición de la inversa, al multiplicar esta fila por la matriz $(I - Q)$, debemos obtener la primera fila de la matriz identidad:
$$\mathbf{n}_0 \cdot (I - Q) = (1, 0, 0, 0, 0, 0, 0)$$

Restamos $I - Q$ y multiplicamos por $\mathbf{n}_0$ para generar un sistema de ecuaciones lineales por columnas:

$n_0 - \frac{1}{2}n_1 - \frac{1}{2}n_2 = 1$  
$-\frac{1}{2}n_0 + n_1 - \frac{1}{4}n_3 = 0$  
$-\frac{1}{2}n_0 + n_2 - \frac{1}{4}n_3 = 0$  
$-\frac{1}{2}n_1 - \frac{1}{2}n_2 + n_3 - \frac{1}{3}n_4 - \frac{1}{3}n_5 = 0$  
$-\frac{1}{4}n_3 + n_4 - \frac{1}{2}n_6 = 0$  
$-\frac{1}{4}n_3 + n_5 - \frac{1}{2}n_6 = 0$  
$-\frac{1}{3}n_4 - \frac{1}{3}n_5 + n_6 = 0$

Aprovechamos la simetría del sistema para resolverlo rápidamente:

De las ecuaciones (2) y (3) es evidente que $n_1 = n_2$.  
De las ecuaciones (5) y (6) es evidente que $n_4 = n_5$.

Sustituyendo $n_4 = n_5$ en la ecuación (7):
$$-\frac{2}{3}n_4 + n_6 = 0 \implies n_6 = \frac{2}{3}n_4$$

Sustituyendo $n_6$ en la ecuación (5):
$$-\frac{1}{4}n_3 + n_4 - \frac{1}{2}\left(\frac{2}{3}n_4\right) = 0$$
$$-\frac{1}{4}n_3 + n_4 - \frac{1}{3}n_4 = 0$$
$$\frac{2}{3}n_4 = \frac{1}{4}n_3 \implies n_4 = \frac{3}{8}n_3$$

Sustituyendo $n_1 = n_2$ y el valor de $n_4, n_5$ en la ecuación (4):
$$-n_1 + n_3 - \frac{2}{3}\left(\frac{3}{8}n_3\right) = 0$$
$$-n_1 + n_3 - \frac{1}{4}n_3 = 0 \implies n_1 = \frac{3}{4}n_3$$

Sustituyendo $n_1$ en la ecuación (2):
$$-\frac{1}{2}n_0 + \left(\frac{3}{4}n_3\right) - \frac{1}{4}n_3 = 0$$
$$-\frac{1}{2}n_0 + \frac{1}{2}n_3 = 0 \implies n_0 = n_3$$

Finalmente, sustituimos $n_0 = n_3$ y $n_1 = n_2 = \frac{3}{4}n_0$ en la ecuación (1):
$$n_0 - \frac{1}{2}\left(\frac{3}{4}n_0\right) - \frac{1}{2}\left(\frac{3}{4}n_0\right) = 1$$
$$n_0 - \frac{3}{4}n_0 = 1 \implies \frac{1}{4}n_0 = 1 \implies \mathbf{n_0 = 4}$$

Sabiendo que $n_0 = 4$ y $n_3 = 4$, podemos encontrar el valor crítico $n_4$ que definimos anteriormente:
$$n_4 = \frac{3}{8}(4) = \frac{3}{2} = 1.5$$

*(Interpretación: Si el ratón inicia en 0, se espera que pase 1.5 veces por la casilla 4 antes de ser absorbido por el sistema).*

##Probabilidad Final de Absorción
Retomamos nuestra ecuación de probabilidad derivada de $B = N \cdot R$:
$$b_{0,7} = n_{0,4} \cdot \frac{1}{3}$$
$$b_{0,7} = \frac{3}{2} \cdot \frac{1}{3}$$
$$b_{0,7} = \frac{1}{2}$$

La probabilidad exacta de que el ratón alcance la comida partiendo de la casilla 0 es de **1/2**.




## Matriz fundamental

La matriz fundamental es:

$$N = (I - Q)^{-1}$$

Luego, la matriz de probabilidades de absorción es:

$$B = N \cdot R$$

Al realizar el cálculo se obtiene:

$$B = \begin{pmatrix}
1/2 & 1/2 \\
2/3 & 1/3 \\
1/3 & 2/3 \\
1/2 & 1/2 \\
2/3 & 1/3 \\
1/3 & 2/3 \\
1/2 & 1/2
\end{pmatrix}$$

## Finalmente, la interpretación final es:

La fila correspondiente al estado inicial $0$ es $\begin{pmatrix} 1/2 & 1/2 \end{pmatrix}$.

Por tanto:

$$P(\text{comida antes que shock} \mid X_0 = 0) = \frac{1}{2} = 0.5$$

Es decir, el ratón tiene una probabilidad del **50%** de encontrar primero la comida antes de recibir el shock.

**SOLUCIÓN ANÁLITICA(MATRIZ FUNDAMENTAL)**

```python
#Librerias utilizadas

import numpy as np# numpy permite trabajar con matrices
# y operaciones de álgebra lineal

#Definición de los estados del problema planteado:

# Estados transitorios:
# el ratón aún no termina
transitorios = [0, 1, 2, 3, 4, 5, 6]

# Definición de los Estados absorbentes:
# 7 = comida
# 8 = shock
absorbentes = [7, 8]
```

```python
#Para Analizar las conexiones del laberinto:

# Se cre un Diccionario que describe
# las posibles salidas desde cada estado
conexiones = {

    # Desde 0 puede ir a 1 o 2
    0: [1, 2],

    # Desde 1 puede ir a:
    # 0, 3 o comida (7)
    1: [0, 3, 7],

    # Desde 2 puede ir a:
    # 0, 3 o shock (8)
    2: [0, 3, 8],

    # Desde 3 puede ir a 4 estados
    3: [1, 2, 4, 5],

    # Desde 4 puede ir a:
    # 3, 6 o comida
    4: [3, 6, 7],

    # Desde 5 puede ir a:
    # 3, 6 o shock
    5: [3, 6, 8],

    # Desde 6 puede ir a 4 o 5
    6: [4, 5],

    # Estados absorbentes
    7: [7],
    8: [8]
}
```

```python
#Para los tamaños de las matrices

# Número de estados transitorios
n = len(transitorios)

# Número de estados absorbentes
m = len(absorbentes)
```

```python
#Se plantean la matriz Q y R

# Matriz Q:
Q = np.zeros((n, n)) #transiciones entre estados transitorios

# Matriz R:
R = np.zeros((n, m)) # transiciones de estados transitorios
# hacia absorbentes
```

```python
#Para construcción de la matriz Q y R:

# Cliclo que recorre cada estado transitorio
for i, estado in enumerate(transitorios):

    # Para Obtener las salidas posibles
    salidas = conexiones[estado]

    # Como el ratón elige al azar,
    # cada salida tiene la misma probabilidad
    prob = 1 / len(salidas)

    # Este ciclo recorre cada destino posible.
    for destino in salidas:

        #Para el caso de que sea transitorio

        if destino in transitorios:

            # Índice del estado destino
            j = transitorios.index(destino)

            # Guardamos la probabilidad
            # en la matriz Q
            Q[i, j] = prob


        #Para el caso de que el destino se absorbente
        if destino in absorbentes:

            # Índice del absorbente
            j = absorbentes.index(destino)

            # Guardamos la probabilidad
            # en la matriz R
            R[i, j] = prob

print("Matriz Q:")
print(Q)

print("\nMatriz R:")
print(R)
```

```python
#Se obtiene Matriz Fundamental

# Matriz identidad
I = np.eye(n)

# Fórmula:
# N = (I - Q)^(-1)
N = np.linalg.inv(I - Q)
print("\nMatriz fundamental N:")
print(N)
```

```python
#Se obtiene Matriz de Absorción

# B = N * R
# Contiene las probabilidades de terminar
# en cada estado absorbente
B = N @ R
print("\nMatriz de absorción B:")
print(B)
```

```python
#Impresión de resultado final

# Probabilidad de llegar a comida
# iniciando desde el estado 0
prob_comida = B[0, 0]
print("\nProbabilidad de llegar a la comida iniciando en 0:")
print(prob_comida)
```

**SOLUCIÓN POR SIMULACIÓN (MONTE CARLO)**

```python
#Librerias utilizadas:
import random# Importamos random para poder elegir movimientos aleatoriosv
import numpy as np# Importamos numpy para calcular el promedio
```

```python
#Se crea un diccionario llamado conexiones. Este diccionario representa el laberinto
# Diccionario que contiene las conexiones del laberinto
# Cada número representa una casilla
# Cada lista representa las casillas a las que se puede mover el ratón
conexiones = {
    0: [1, 2],        # Desde la casilla 0 puede ir a 1 o 2
    1: [0, 3, 7],     # Desde la casilla 1 puede ir a 0, 3 o comida
    2: [0, 3, 8],     # Desde la casilla 2 puede ir a 0, 3 o shock
    3: [1, 2, 4, 5],  # Desde la casilla 3 puede ir a 1, 2, 4 o 5
    4: [3, 6, 7],     # Desde la casilla 4 puede ir a 3, 6 o comida
    5: [3, 6, 8],     # Desde la casilla 5 puede ir a 3, 6 o shock
    6: [4, 5]         # Desde la casilla 6 puede ir a 4 o 5
}
```

```python
#Se crea una función llamada raton. Esta función simula una caminata completa del ratón desde la casilla 0 hasta que llegue a comida o shock.
# Definimos una función que simula un recorrido completo del ratón
def raton():

    # El ratón inicia en la casilla 0
    posicion = 0

    # Mientras el ratón no llegue a comida ni a shock, seguirá moviéndose
    while posicion not in [7, 8]:

        # Se Elige al azar una de las posibles salidas desde su posición actual
        posicion = random.choice(conexiones[posicion])

    # Regresa True si llegó a la comida
    # Regresa False si llegó al shock
    return posicion == 7
```

```python
# Número total de simulaciones
N = 100000

# Se simula el recorrido del ratón N veces
# Cada True cuenta como 1 y cada False como 0
# Por eso, el promedio representa la probabilidad de llegar a la comida
prob = np.mean([raton() for _ in range(N)])

print("Probabilidad de llegar a la comida:")
# Se Imprimime la probabilidad simulada
print(prob)
```
